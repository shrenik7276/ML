```python
#principle components
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Iris.csv")
```

```python
list=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
```

```python
x=df[list]
y=df['Species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size=0.8,random_state=32)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
from sklearn.decomposition import PCA
```

```python
p=PCA(n_components=1)
```

```python
A=p.fit_transform(x_train)
```

```python
B=p.transform(x_test)
```

```python
A
```

```python
B
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=15)
```

```python
knn.fit(A,y_train)
```

```python
pre=knn.predict(B)
```
