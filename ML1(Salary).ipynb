```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('salary_data.csv')
```

```python
x=np.array(df.iloc[0:25,0]).reshape(-1,1)
```

```python
y=np.array(df.iloc[0:25,1]).reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size=0.8,random_state=32)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR=LinearRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
pre=LR.predict(x_test)
```

```python
print(pre)
```

```python
print(y_test)
```

```python
LR.intercept_
```

```python
LR.coef_
```

```python
plt.scatter(x,y)
plt.plot([min(x),max(x)],[min(pre),max(pre)],color='red')
plt.show()
```

```python
from sklearn import metrics
metrics.mean_absolute_error(y_test,pre)
metrics.mean_squared_error(y_test,pre)
```
